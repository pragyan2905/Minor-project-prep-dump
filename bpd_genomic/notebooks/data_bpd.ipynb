**Soft file into a csv**

```python
import pandas as pd
import re
import io

soft_file = "GDS3356_full.soft 2"

with open(soft_file, 'r', encoding='latin1') as f:
    lines = f.readlines()

start = None
end = None

for i, line in enumerate(lines):
    if "!dataset_table_begin" in line:
        start = i + 1
    if "!dataset_table_end" in line:
        end = i
        break

if start is None or end is None:
    raise ValueError("Could not locate table section in SOFT file.")

table_lines = lines[start:end]
genomic_df = pd.read_csv(io.StringIO("".join(table_lines)), sep="\t")
genomic_df.columns = [c.replace('"', '') for c in genomic_df.columns]

genomic_df.to_csv("genomic_nbp.csv", index=False)

print("Saved genomic_nbp.csv successfully!")
print("Shape:", genomic_df.shape)
```

**Txt file into csv**

```python
import pandas as pd
import re

series_file = "GSE8586_series_matrix.txt"

sample_ids = []
titles = []

with open(series_file, "r", encoding="latin1") as f:
    for line in f:
        line = line.strip()

        if line.startswith("!Sample_geo_accession"):
            parts = line.split("\t")[1:]
            sample_ids = [p.replace('"', '') for p in parts]


        if line.startswith("!Sample_title"):
            parts = line.split("\t")[1:]
            titles = [p.replace('"', '') for p in parts]

labels = []
for t in titles:
    if "bpd" in t.lower():
        labels.append("bpd")
    else:
        labels.append("nobpd")

pheno_df = pd.DataFrame({
    "Sample_ID": sample_ids,
    "Label": labels,
    "Original_Title": titles
})

pheno_df.to_csv("phenotype_labels.csv", index=False)

print("Saved phenotype_labels.csv successfully!")
print(pheno_df.head())
```
